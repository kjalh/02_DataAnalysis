# **1. Series**
Series는 Pandas에서 사용하는 가장 기본적인 1차원 데이터 구조로, 값(Value)과 인덱스(Index)가 함께 구성된 자료형입니다. 
<br>NumPy의 1차원 배열과 비슷하지만, 각 데이터에 이름표 역할을 하는 인덱스가 추가되어 데이터를 더 체계적으로 관리할 수 있습니다.

![시리즈](./images/시리즈.png)

````python
idx = ['김사과', '반하나', '오렌지', '이메론', '배애리']
data = [67, 75, 90, 62, 98]

# pd.Series(데이터, 인덱스, ...)
pd.Series(data)
````

````python
ser1 = pd.Series(data, idx)
print(ser1)
print(ser1.index)
print(ser1.values)
print(type(ser1.values))
````

### Index 객체란
Index 객체는 Series와 DataFrame에서 각 데이터의 위치를 식별하고 관리하기 위한 “이름표(Label)” 역할을 하는 자료구조입니다. 
<br>일반 리스트처럼 보이지만 단순한 데이터 저장용이 아니라, 데이터를 빠르게 검색하고 정렬하며 서로 다른 데이터끼리 매칭할 수 있도록 최적화된 Pandas 전용 객체입니다.

![인덱스 객체](./images/인덱스객체.png)

# **2. DataFrame**
데이터프레임(DataFrame)은 판다스(Pandas) 라이브러리에서 제공하는 중요하고 강력한 데이터 구조로, 2차원의 테이블 형태 데이터를 다루는 데 사용됩니다. 
<br>또한 각 요소는 인덱스(index), 열(column), 값(value)으로 구성되어 있습니다. 데이터프레임은 행과 열로 이루어져 있으며, 각 열은 다양한 데이터 타입을 가질 수 있습니다. 
<br>값은 넘파이의 ndarray 기반으로 저장됩니다.

![데이터 프레임](./images/데이터프레임.png)

````python
data = [[67, 93, 91],
        [75, 68, 96],
        [87, 81, 82],
        [62, 70, 75],
        [98, 56, 87]]


idx = ['김사과', '반하나', '오렌지', '이메론', '배애리']
col = ['국어', '영어', '수학']



# pd.DataFrame(데이터, 인덱스, 컬럼, ...)
pd.DataFrame(data)
````

````python
pd.DataFrame(data, idx)
````

````python
pd.DataFrame(data, idx, col)
````

````python
df = pd.DataFrame(index=idx, columns=col, data=data)  # 순서는 달라도 크게 달라지는 건 없음
df
````

````python
print(df.index)
print(df.columns)
print(df.values)
````

````python
# 딕셔너리를 사용하여 데이터프레임을 생성하기
dic = {
    '국어':[67, 75, 76, 62, 98],
    '영어':[93, 68, 81, 70, 56],
    '수학':[91, 96, 82, 75, 87]
}
df = pd.DataFrame(data=dic, index=idx)
df
````

# **4. 정렬하기**

````python
df.sort_index() # index로 오름차순 정렬: 기본값
````

````python
df.sort_index(ascending=False) # index로 내림차순 정렬
````

````python
df.sort_values(by='height') # 키로 오름차순 정렬
````

````python
df.sort_values(by='height', ascending=False) # 키로 내림차순 정렬

# 👉  문자열 비교는 사전(Dictionary) 순서처럼 비교하기 때문에 "182cm" > "182.2cm" 처럼 이상한 결과가 나올 수 있습니다.
# 여기서는 .보다 c가 아스키 값이 더 크기 때문에 182.2cm이 작게 나옴
````

````python
df['height'] = df['height'].str.replace('cm', '').astype(float)
df
# 👉   cm 문자열을 제거한 뒤 숫자(float)로 변환
````

````python
# 진짜로 바뀐지 확인
df.info()
````

````python
df['brand'] = df['brand'].str.replace(',', '').astype(int)
df
# 👉    brand 컬럼도 현재 숫자가 아니라 문자열(String) 상태. 쉼표(,)를 제거한 뒤 숫자로 변환
````

````python
df.info()
````

````python
# loc 인덱싱
df.loc[:, 'name'] # df['name'] # 왼쪽은 행 오른쪽은 열
````

> 👉 loc는 DataFrame이나 Series에서 인덱스 이름(Label)과 컬럼명을 기준으로 데이터를 선택하는 인덱서(Indexer)입니다. 
<br> 즉, 숫자 위치가 아니라 실제 행 이름과 열 이름을 사용하여 데이터를 가져옵니다. loc는 사람이 읽기 쉽고 데이터 의미를 기준으로 접근하기 때문에 데이터 분석과 전처리 작업에서 가장 많이 사용하는 방식 중 하나입니다.

````python
df.loc[2:5, 'name']  # 2부터 5까지 (인덱스를 사용하는게 아니라 인덱스에 들어있는 숫자를 사용. 즉, 인덱스 안에 있는 값을 사용)
````

````python
df.loc[2:5, ['name', 'gender', 'height']]
````

````python
df.loc[[2,5], ['name', 'gender', 'height']]
````

````python
df.loc[2:5, 'name':'gender']
````

````python
# iloc 인덱싱: index로 인덱싱
df.iloc[:, 0]
````

> 👉 iloc는 DataFrame이나 Series에서 숫자 위치(Index Position)를 기준으로 데이터를 선택하는 인덱서(Indexer)입니다. 
<br> 즉, 실제 인덱스 이름이 아니라 행과 열의 순서 번호를 사용하여 데이터를 가져옵니다. iloc는 데이터의 실제 이름과 상관없이 순서 기준으로 접근하기 때문에 반복 처리나 특정 위치 데이터를 다룰 때 많이 사용됩니다.

````python
df.iloc[:, 0:3] # 3을 포함하지 않음 # iloc는 인덱스 사용 그래서 0 1 2 즉, 0부터 3개 뽑아라
````

````python
df.iloc[:, [0, 3]]
````

````python
df.iloc[1:5, 0:2]
````

````python
df['height'] >= 180 # height 있는 모든 값들과 비교
````

````python
df[df['height'] >= 180] # 180이 넘는 모든 값  SELECT * FROM Table WHERE height>= 180
````

````python
df[df['height'] >= 180] # SELECT name FROM Table WHERE height >= 180
````

````python
df[df['height'] >= 180][['name', 'gender', 'height']] # SELECT name, gender, height FROM Table  WHERE height >= 180
````

````python
df.loc[df['height'] >= 180, ['name', 'gender', 'height']] # SELECT name, gender, height FROM Table  WHERE height >= 180
````

````python
# df.loc[df['height'] >= 180, ['name', 'gender', 'height']]와 같음
df.loc[df['height'] >= 180, ['name', 'gender', 'height']]   # SELECT name, gender, height FROM Table  WHERE height >= 180
````
