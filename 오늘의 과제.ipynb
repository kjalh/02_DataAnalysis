```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def melon_artist_tracks_all_pages():
    headers = {
        "User-Agent": "Mozilla/5.0",
        "Referer": "https://www.melon.com/index.htm"
    }

    all_data = []

    
    url = (
        f"https://www.melon.com/chart/index.htm"
    )
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    rows= soup.select("tbody tr")

    # print(len(rows)) # 가져오는지 증명
    
    page_data = []

    for row in rows:
        rank_tag = row.select_one("span.rank")
        title_tag = row.select_one("div.rank01 a")
        artist_tag = row.select_one("div.rank02 a")
        album_tag = row.select_one("div.rank03 a")
        like_tag = row.select_one("span.cnt text")

        if title_tag is None:
            continue
        
        page_data.append({
            "순위": rank_tag.get_text(strip=True),
            "곡명": title_tag.get_text(strip=True),
            "아티스트": artist_tag.get_text(strip=True) if artist_tag else "",
            "앨범명": album_tag.get_text(strip=True) if album_tag else "",
            "좋아요 수": like_tag.get_text(strip=True).replace("총건수","") if like_tag else ""
        })

        if len(page_data) == 0:
            # print(f"{page}100곡 끝 일단해")
            break
        
        
        print(f"크롤링 완료: {len(page_data)}곡")
    all_data.extend(page_data)
    df = pd.DataFrame(all_data)

    if not df.empty:
        df = df.drop_duplicates(subset=["순위", "곡명", "아티스트", "앨범명", "좋아요 수"])
        df.index = "" # df.index = df.index + 1

    file_name = f"melon_top100_tracks.csv"
    df.to_csv(file_name, encoding="utf-8-sig", index = False)

    print(f"CSV 저장 완료: {file_name}")
    print(f"총 {len(df)}곡 수집 완료")

    return df
        
```

```python
df = melon_artist_tracks_all_pages()
df
```
