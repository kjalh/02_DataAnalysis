# 좋아요 수 안 나옴

```python
def melon_artist_tracks_all_pages():
    headers = {
        "User-Agent": "Mozilla/5.0",
        "Referer": "https://www.melon.com/index.htm"
    }

    all_data = []

    
    url = (
        f"https://www.melon.com/chart/index.htm"
    )
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    rows= soup.select("tbody tr")

    # print(len(rows)) # 가져오는지 증명
    
    page_data = []

    for row in rows:
        rank_tag = row.select_one("span.rank")
        title_tag = row.select_one("div.rank01 a")
        artist_tag = row.select_one("div.rank02 a")
        album_tag = row.select_one("div.rank03 a")
        like_tag = row.select_one("span.cnt")

        print(repr(like_tag.get_text(strip=True)))
        
        if title_tag is None:
            continue

        
        page_data.append({
            "순위": rank_tag.get_text(strip=True),
            "곡명": title_tag.get_text(strip=True),
            "아티스트": artist_tag.get_text(strip=True) if artist_tag else "",
            "앨범명": album_tag.get_text(strip=True) if album_tag else "",
            "좋아요 수": like_tag.get_text(strip=True).replace("총건수", "") if like_tag else ""
        })

        if len(page_data) == 0:
            # print(f"{page}100곡 끝 일단해")
            break
        
        
        print(f"크롤링 완료: {len(page_data)}곡")
    all_data.extend(page_data)
    df = pd.DataFrame(all_data)

    if not df.empty:
        df = df.drop_duplicates(subset=["순위", "곡명", "아티스트", "앨범명", "좋아요 수"])
        # df.index = df.index + 1

    file_name = f"melon_top100_tracks.csv"
    df.to_csv(file_name, encoding="utf-8-sig", index =False)

    print(f"CSV 저장 완료: {file_name}")
    print(f"총 {len(df)}곡 수집 완료")

    return df
        
```

```python
df = melon_artist_tracks_all_pages()
df
```

# 또 다른 방법

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def melon_artist_tracks_all_pages():
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36",
        "Referer": "https://www.melon.com/index.htm"
    }

    all_data = []

    url = f"https://www.melon.com/chart/index.htm"
    response = requests.get(url, headers=headers)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")
    rows = soup.select("tbody tr")

    # =========================================================
    # [추가된 API 로직] 반복문을 돌기 전에 미리 100곡의 정답지를 받아옵니다.
    song_ids = [row.select_one("button.like")['data-song-no'] for row in rows if row.select_one("button.like")]
    like_api_url = f"https://www.melon.com/commonlike/getSongLike.json?contsIds={','.join(song_ids)}"
    like_data = requests.get(like_api_url, headers=headers).json()
    # =========================================================

    page_data = []

    # [수정된 부분] zip을 써서 HTML(row)과 정답지(like)를 동시에 꺼냅니다.
    for row, like in zip(rows, like_data['contsLike']):
        rank_tag = row.select_one("span.rank")
        title_tag = row.select_one("div.rank01 a")
        artist_tag = row.select_one("div.rank02 a")
        album_tag = row.select_one("div.rank03 a")

        if title_tag is None:
            continue

        page_data.append({
            "순위": rank_tag.get_text(strip=True),
            "곡명": title_tag.get_text(strip=True),
            "아티스트": artist_tag.get_text(strip=True) if artist_tag else "",
            "앨범명": album_tag.get_text(strip=True) if album_tag else "",
            "좋아요 수": f"{like['SUMMCNT']:,}" # 가짜 0 대신 진짜 좋아요 숫자를 넣습니다!
        })

        if len(page_data) == 0:
            break
        
        print(f"크롤링 진행 중: {len(page_data)}곡")

    all_data.extend(page_data)
    df = pd.DataFrame(all_data)

    if not df.empty:
        df = df.drop_duplicates(subset=["순위", "곡명", "아티스트", "앨범명", "좋아요 수"])

    file_name = f"melon_top100_tracks.csv"
    df.to_csv(file_name, encoding="utf-8-sig", index=False)

    print(f"CSV 저장 완료: {file_name}")
    print(f"총 {len(df)}곡 수집 완료")

    return df

melon_artist_tracks_all_pages()
```
